Discover and Visualize the Data to Gain Insights

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(10,6))
data.groupby('year')['tourism_arrivals'].sum().plot()
plt.title('Global Tourism Arrivals Over Time')
plt.xlabel('Year')
plt.ylabel('Tourism Arrivals')
plt.grid(True)
plt.show()
```

```python
data.plot(kind="scatter", x="tourism_arrivals", y="gdp", alpha=0.2)
plt.title('Tourism Arrival vs GDP')
```

```python
data.plot(kind="scatter", x="tourism_arrivals", y="gdp", alpha=0.7,
             s=data["unemployment"]/100, label="year", figsize=(8,4),
             c="inflation", cmap=plt.get_cmap("jet"), colorbar=True,
             sharex=False)
plt.legend()
```

Looking for Correlation

```python
import seaborn as sns
dt = data[['gdp', 'unemployment','tourism_departures', 'tourism_arrivals' ]]
print(dt)
```

```python
dt.corr()
```

```python
plt.figure(figsize=(16, 6))
heatmap = sns.heatmap(dt.corr(), vmin=-1, vmax=1, annot=True, cmap='BrBG')
heatmap.set_title('Correlation Heatmap', fontdict={'fontsize':18}, pad=12);
```

```python
sns.pairplot(dt, diag_kind='kde', plot_kws={'alpha': 0.6})
plt.suptitle("Scatter Plot", y=1.02)
plt.show()
```

Insight From Data

Global tourism trends consistently increased from 2000 to 2018, with the number of tourist arrivals steadily rising from around 10 billion to 20 billion. However, 2020 recorded a sharp decline that almost reached zero, most likely due to the COVID-19 pandemic. The positive correlation that can be seen in the scatter plot between the number of tourist arrivals and departures and GDP indicates that countries with stronger economies tend to have more developed tourism industries. Conversely, the shrinking blue dots in the scatter plot indicate that countries with strong economies and developed tourism industries tend to have lower unemployment and inflation rates. This is reinforced by the scatter plot and correlation heatmap, which show that countries with high GDP have a positive correlation with tourism activities, such as tourism departures and tourism arrivals, characterized by correlation values close to 1. Conversely, countries with high GDP have a negative correlation with unemployment and inflation rates, indicated by negative correlation values below 0. Therefore, countries need to continue developing tourism infrastructure to ensure stability and sustainable economic growth.
